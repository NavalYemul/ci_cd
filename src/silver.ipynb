```python
from pyspark.sql.functions import *
```

```python
dbutils.widgets.text("catalog","develop")
```

```python
catalog=dbutils.widgets.get("catalog")
```

```python
df=spark.table(f"{catalog}.bronze.sales")
```

```python
df1=df.dropDuplicates().dropna().drop("ingestion_date","path")
```

```python
df1.write.mode("overwrite").saveAsTable(f"{catalog}.silver.sales_cleaned")
```

```python
df=spark.table(f"{catalog}.silver.sales_cleaned")
```

```python
df1=df.groupBy("product_id").agg(sum("total_amount").alias("total_sales"))
```

```python
df1.write.mode("overwrite").saveAsTable(f"{catalog}.gold.total_sales_product")
```
