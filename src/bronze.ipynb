```python
from pyspark.sql.functions import *
```

```python
dbutils.widgets.text("catalog","develop")
dbutils.widgets.text("schema","bronze")
```

```python
catalog = dbutils.widgets.get("catalog")
schema = dbutils.widgets.get("schema")

base_path = f"/Volumes/{catalog}/bronze/raw"

df = spark.read.csv(f"{base_path}/sales.csv", header=True, inferSchema=True)
```

```python
df1=df\
    .withColumn("ingestion_date",current_timestamp())\
    .withColumn("path",col("_metadata.file_path"))
```

```python
df1.write.mode("overwrite").saveAsTable(f"{catalog}.bronze.sales")
```
